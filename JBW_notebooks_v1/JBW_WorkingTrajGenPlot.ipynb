```python
from __future__ import print_function
from mpl_toolkits.basemap import Basemap
import pysplit
import matplotlib.pyplot as plt
import numpy as np
```

```python
trajgroup = pysplit.make_trajectorygroup(r'C:/hysplit4/sumatra/*dec*')
print('Total trajectories in group: ',trajgroup.trajcount)
```

```python
fig = plt.figure()

bmap = pysplit.mapdesigner.Basemap(llcrnrlon = 60,
llcrnrlat = -30,
urcrnrlon = 130,
urcrnrlat = 30,
projection = 'cyl')

bmap.drawmeridians(np.arange(int(60),int(140),10),labels=[0,0,0,1],linewidth=0.5)
bmap.drawparallels(np.arange(int(-20),int(60),5),labels=[1,0,0,0],linewidth=0.5)
bmap.drawstates()
bmap.drawcoastlines()
bmap.drawcountries()

bmap.drawmapboundary(color='k',linewidth=1.0,fill_color='0.90')
bmap.fillcontinents(color='0.80', lake_color=None)

for traj in trajgroup:
    bmap.plot(*traj.path.xy, c='blue', latlon=True, zorder=20)

plt.show()
```

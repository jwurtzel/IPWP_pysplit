```python
# STEP 2: INITIALISE TRAJECTORY GROUP
# Group trajectories and then cycle through a group to determine distance, vector and rain status.
# check out traj_trajgroup_basics_examle.py

# Initialise Trajectories and Create a Trajectory Group
WJtrajroup=[]
WJtrajgroup = py.make_trajectorygroup(r'/volumes/NickScroxtonUMass/traj_dir/WeeJasper/WeeJasperoct*')
# Calculate distance and vectors
for traj in WJtrajgroup:
    traj.calculate_distance()
    traj.calculate_vector()
print('Total trajectories in group: ',WJtrajgroup.trajcount)
```

```python
# STEP 3A: CALCULATE INTEGRATION ERRORS
# Calculate integration errors using reverse trajectories
# check out integration_error.py
c = 0
for traj in WJtrajgroup[:]:
    traj.load_reversetraj()
    traj.calculate_integrationerr()
    print(c)
    c+=1
```

```python
# DEBUGGING STEP FOR IF MOISTURE UPTAKE AND HYPATH FAILS:
# FIND DISRUPTIVE PATH, REMOVE IT, RETURN TO STEP 2
print('Error at trajectory',c,'of',len(list(enumerate(WJtrajgroup))))
pprint (vars(WJtrajgroup[c]))
pprint (vars(WJtrajgroup[c-1]))
```

```python
# STEP 3B: Determine Cutoff and apply to remove bad trajectories
# check out integration_error.py

relative_errors=[traj.integration_error for traj in WJtrajgroup]
# Determine what qualifies as a bad trajory
cutoff=np.mean(relative_errors)+np.std(relative_errors)*2
print('Integration error upper limit: ', cutoff)
# Calculate the bad trajectories
good_traj=[]
for traj in WJtrajgroup:
    if traj.integration_error < cutoff:
        good_traj.append(traj)
good_traj=py.TrajectoryGroup(good_traj)        
print('Result: ', good_traj.trajcount, 'trajectories')
```

```python
# Calculate rain status
rainylist=[]
for traj in good_traj:
    traj.set_rainstatus()
    if traj.rainy:
        rainylist.append(traj)
rainy_trajgroup = py.TrajectoryGroup(rainylist)
dry_trajgroup=good_traj-rainy_trajgroup
print('Total trajectories in group: ',WJtrajgroup.trajcount)
print('Good trajectories: ',good_traj.trajcount)
print('Dry trajectories: ',dry_trajgroup.trajcount)
print('Rainy trajectories: ',rainy_trajgroup.trajcount)
```

```python
#Additional Calculations
for traj in rainy_trajgroup:
    traj.calculate_vector()
    traj.calculate_sh()
    traj.calculate_moistureflux()
```
